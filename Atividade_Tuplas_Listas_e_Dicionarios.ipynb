# **Exercícios Tuplas e Listas**

```python
'''
TAREFA 1

Escreva um programa que solicita que o usuário digite 5 números inteiros e os
coloca em uma lista.

Quando terminar, imprima a lista.
'''

lista = []
print('Digite 5 números INTEIROS a seguir:')

for i in range(5):
    numero = int(input(f'{i + 1}° número: '))
    lista.append(numero)

print(lista)
```

```python
'''
TAREFA 2

Escreva um programa que aplica a função módulo a todos os elementos em uma
lista.
'''

lista1 = [1, -20, 9, -51, 27]

for i in range(len(lista1)):
  lista1[i] = abs(lista1[i])

print(lista1)
```

```python
'''
TAREFA 3

Escreva um programa que apaga todos os elementos negativos de uma lista.
'''

lista = [3, -5, -11, 12, -50, -90]

for i in range(len(lista) - 1, -1, -1):
  if lista[i] < 0:
    lista.pop(i)

print(lista)
```

```python
'''
TAREFA 4

Faça um programa que solicita uma quantidade indefinida de notas de uma prova
(deve parar quando um número negativo for digitado).

Armazena em uma lista.

Após isso, o programa deve mostrar a média
das notas.
'''

notas = []
print('Caso uma nota negativa for digitada, o sistema irá parar.')

while True:
  nota = float(input('Digite uma nota: '))
  print()

  if nota < 0: break

  notas.append(nota)

media = sum(notas) / len(notas)
print(f'A média das notas é: {media:.2f}')
```

```python
'''
TAREFA 5

Faça um programa que solicita uma quantidade indefinida de números positivos
(deve parar quando um número negativo for digitado).

Armazene-os em duas listas:
Uma para os números pares; outra para os números ímpares.

A seguir, mostre a porcentagem de pares e ímpares digitados.
'''

par = []
impar = []
total = 0

print('Apenas números positivos.')
print('Caso um número negativo for digitado, o sistema irá parar.')

while True:
  num = int(input('Digite um número: '))
  print()

  if num < 0: break

  if num % 2 == 0:
    par.append(num)
    total += 1
  else:
    impar.append(num)
    total += 1

print('\nPorcentagem de números pares digitados: ', (len(par) / total) * 100)
print('Porcentagem de números ímpares digitados: ', (len(impar) / total) * 100)
```

```python
'''
TAREFA 6
Faça um programa que solicita uma quantidade indefinida de números positivos
(deve parar quando um número negativo for digitado).

Armazene todos os números digitados em uma lista, sem repetição.
'''

lista = []

print('Caso um número negativo for digitado, o sistema irá parar.')

while True:
  num = int(input('Digite um número positivo: '))
  print()

  if num < 0: break

  if num not in lista:
    lista.append(num)

print(lista)
```

```python
'''
TAREFA 7
Escreva um programa que solicita que o usuário digite a altura e o peso de
5 pessoas.

Verifique se pelo menos duas das pessoas tem a mesma altura e mesmo peso.
'''

pessoas = []

print('A seguir, digite a altura e o peso de 5 pessoas.')

for i in range(5):
  altura = float(input(f'Digite a altura da {i + 1}° pessoa: '))
  peso = float(input(f'Digite o peso da {i + 1}° pessoa: '))
  pessoas.append((altura, peso))

iguais = 0

for i in range(len(pessoas) - 1):
  if pessoas[i] == pessoas[i + 1]:
    iguais += 1

if iguais >= 1:
  print('Das informações digitadas. Há pessoas que tem a mesma altura e peso.')
else:
  print('Nenhuma das pessoas tem a mesma altura e peso.')
```

```python
'''
TAREFA 8

Faça um programa que solicita que o usuário digite o nome e a idade de
diversas pessoas (o programa deve parar quando um nome vazio for informado).

Retorne o nome da pessoa mais velha.
'''

pessoas = []
maior = 0
nome_maior = ''
cont = 1

print('Digite o nome e a idade de várias pessoas.')
print('Caso um nome vazio for inserido, o sistema será encerrado.')
print()

while True:
  nome = input(f'Digite o nome da {cont}° pessoa: ')

  if nome == '': break

  idade = int(input('Insira a idade dessa pessoa: '))

  if maior <= idade:
    maior = idade
    nome_maior = nome
  print()

  pessoas.append((nome, idade))
  cont += 1

if cont == 1:
  print('A pessoa mais velha é: ninguém!')
else:
  print('A pessoa mais velha é:', nome_maior)
```

```python
'''
TAREFA 2
Dado o dicionário abaixo:

animes = {
  "Naruto": 220,
  "Jujutsu Kaisen": 47,
  "Dragon Ball Z": 291,
  "Death Note": 37,
  "Fullmetal Alchemist": 64,
  "Evangelion": 26,
  "Berserk": 25,
  "Code Geass": 50,
  "Akame ga Kill!": 24,
  "Elfen Lied": 13
}

1 - Imprima o total de episódios
2 - Substitua a quantidade de episódios pelo percentual (ep_anime/ep_total)
3 - Imprima nome e percentual de cada anime, em ordem decrescente de
percentual
'''

animes = {
  "Naruto": 220,
  "Jujutsu Kaisen": 47,
  "Dragon Ball Z": 291,
  "Death Note": 37,
  "Fullmetal Alchemist": 64,
  "Evangelion": 26,
  "Berserk": 25,
  "Code Geass": 50,
  "Akame ga Kill!": 24,
  "Elfen Lied": 13
}

ep_total = sum(animes.values())
ep_anime = len(animes)

print('Total de episódios:', ep_total)
print()
print('Percentual:', (ep_anime / ep_total) * 100)
print()

for anime, ep in sorted(animes.items(), key=lambda anime: anime[1],
                        reverse = True):
  print(f'{anime}: {(ep / ep_total) * 100:.2f}%')
```
