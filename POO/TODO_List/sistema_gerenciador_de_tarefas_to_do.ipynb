# Sistema de Gerenciamento de Tarefas

```python
# Criação de um sistema TO-DO List com CRUD em Memória

# Criação da lista vazia
tarefas = []


# Início da função "add_tarefa"
def add_tarefa():
  # Inserir nome da tarefa
  tarefa = input('Digite a nova tarefa: ')

  while True:
    print('\nPrioridades:')
    print('1 - Alta')
    print('2 - Média')
    print('3 - Baixa')
    print()

    prioridade = input('Escolha o nível de prioridade dessa tarefa: ')

    if prioridade == '1':
      prioridade = 'Alta'
      break
    elif prioridade == '2':
      prioridade = 'Média'
      break
    elif prioridade == '3':
      prioridade = 'Baixa'
      break
    else:
      print('Prioridade inserida é inválida. Insira novamente.')

  id = len(tarefas) + 1

  tarefas.append({
      'id': id,
      'tarefa': tarefa,
      'prioridade': prioridade,
      'concluida': False
  })

  print(f'\nFoi adicionada a tarefa: {tarefa}')
  print('O ID dessa tarefa será:', id)

# Fim da função "add_tarefa"


# Início da função "del_tarefa"
def del_tarefa():
  if len(tarefas) == 0:
    print('Não há tarefas cadastradas.')
    return

  id = int(input('Digite o ID da tarefa que deseja excluir: '))

  for tarefa in tarefas:
    if tarefa['id'] == id:
      tarefas.remove(tarefa)
      print(f'A tarefa {tarefa["tarefa"]} foi excluída.')
      return
  print('Tarefa não encontrada.')

# Fim da função "del_tarefa"


# Início da função "mudar_status"
def mudar_status():
  if len(tarefas) == 0:
    print('Não há tarefas cadastradas.')
    return

  id = int(input('Digite o ID da tarefa que deseja mudar o status: '))

  for tarefa in tarefas:
    if tarefa['id'] == id:
      print('A tarefa se encontra como:', 'Concluída' if tarefa['concluida'] else 'Pendente')
      print('\nEscolha o novo status para essa tarefa:')
      print('1 - Tarefa concluída')
      print('2 - Tarefa pendente')
      print()

      status = input('Digite a opção desejada: ')

      if status == '1':
        tarefa['concluida'] = True
        print(f'A tarefa {tarefa["tarefa"]} foi marcada como concluída.')
      elif status == '2':
        tarefa['concluida'] = False
        print(f'A tarefa {tarefa["tarefa"]} foi marcada como pendente.')
      else:
        print('Opção inserida não é válida. Insira novamente.')
      return
  print('Tarefa não encontrada.')


# Fim da função "mudar_status"


# Início da função "listar_tarefas"
def listar_tarefas(tarefas):
  if len(tarefas) == 0:
    print('Não há tarefas cadastradas.')

  for tarefa in tarefas:
    status = 'CONCLUÍDA' if tarefa['concluida'] else 'PENDENTE'
    print(f'Tarefa: {tarefa['tarefa']} | ID: {tarefa['id']} | Prioridade: {tarefa['prioridade']:5s}| Status: {status}')

# Fim da função "listar_tarefas"


# Primeira saída apresentada
print('--Bem-Vindo(a) ao Gerenciador de Tarefas--')

while True:
  print('\n1 - Adicionar nova tarefa')
  print('2 - Excluir tarefa')
  print('3 - Mudar status da tarefa')
  print('4 - Listar tarefas')
  print('5 - Sair')
  print()

  opcao = input('Digite a opção desejada: ')
  print()

  if opcao == '1':
    add_tarefa()
  elif opcao == '2':
    del_tarefa()
  elif opcao == '3':
    mudar_status()
  elif opcao == '4':
    prioridade_ordem = {'Alta': 1, 'Média': 2, 'Baixa': 3}

    tarefas_ordem = sorted(
        tarefas, key=lambda tarefa: prioridade_ordem[tarefa['prioridade']]
    )

    listar_tarefas(tarefas_ordem)
  elif opcao == '5':
    break
  else:
    print('Opção inserida não é válida. Insira novamente.')


# Resultado da aplicação exibido após a seleção da opção "5"

print('Informações atuais:')
listar_tarefas(tarefas_ordem)
print('\nPrograma encerrado, volte sempre!')
```
